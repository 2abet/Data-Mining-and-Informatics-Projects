```python
import csv
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
time.sleep(2)
import warnings
warnings.filterwarnings('ignore')
```

```python
urls = []
url1 = 'https://www.imdb.com/title/tt6966692/reviews?sort=curated&dir=desc&ratingFilter=3'
url2 = 'https://www.imdb.com/title/tt0198781/reviews?sort=curated&dir=desc&ratingFilter=3'
url3 = 'https://www.imdb.com/title/tt0222012/reviews/?ref_=tt_ov_rt'
url4 = 'https://www.imdb.com/title/tt0112870/reviews?sort=curated&dir=desc&ratingFilter=3'
url5 = 'https://www.imdb.com/title/tt0167260/reviews/?ref_=tt_ov_rt'
url6 = 'https://www.imdb.com/title/tt0270846/reviews/?ref_=tt_ov_rt'
url7 = 'https://www.imdb.com/title/tt2631186/reviews/?ref_=tt_ov_rt'
```

```python
urls.append(url1)
urls.append(url2)
urls.append(url3)
urls.append(url4)
urls.append(url5)
urls.append(url6)
urls.append(url7)
```

```python
content = []
for url in urls:
    page = requests.get(url, timeout=2.50)
    page_content = page.content
    soup = BeautifulSoup(page_content, 'html.parser')
    content.append(soup.find_all('div', class_= 'review-container'))
```

```python
print(content)
```

```python
movie = pd.DataFrame(columns=['Review','Rating'])
```

```python
review = []
rating = []
count = 0
for cc in content:
    for c in cc:    
        count+= 1   
    
        print('\nMovie review ', count)
        #Get review.
        str = c.find_all('a', attrs={'class':'title'})      
        rReview =''
        for s in str:
            #print('Review is: ',s.get_text())
            rReview = s.get_text()
      
        #Get rating.
        ratings = c.find_all('span', attrs={'class':''})
        rVal = []
        for r in ratings:
            str1 = r.get_text().strip()
            rVal.append(str1)
    
        val = rVal[0]
        if(len(val) > 2):
            continue     
        else:
            review.append(rReview)
            rating.append(val)
            print('Review: ', rReview)
            print('Rating: ',val) 
        
movie['Review'] = review
movie['Rating'] = rating
```

```python
import string
import re
#import nltk
#nltk.download()
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
textFeatures = movie['Review'].copy()
textFeatures.shape
```

```python
#Preparing text for Wordcloud
text = []
for t in textFeatures:
  text.append(t)
all_text = ', '.join(t for t in text)
#print(all_text)
print(len(all_text))
```

```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Create stopword list
stopwords = set(STOPWORDS)
stopwords.update(["br", "im", "thats"]) #"im","lol","Xa","film"])
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(all_text)
# Display the image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
#save the generated image to a file
#wordcloud.to_file("wordcloud_cb_all.png")
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
c = 0
for t in text:
    c+=1
    print(c, t)
    ss = sid.polarity_scores(t)
    print(ss)
    
    if(ss['compound'] >= 0.05):
        print('positive')
 
    elif(ss['compound'] <= -0.05):
        print('negative')
    else:
        print('neutral')
    print('\n')
```

```python
label = []
for r in movie['Rating']:
    r = int(r)
    if (r>5):
        label.append('1') #Positive
    elif(r<5):
        label.append('-1') #Negative
    elif(r==5):
        label.append('0') #Netural
movie['class-label'] = label
```

```python
import nltk
nltk.download('punkt')
# Stemming using TextBlob library for stemming
from textblob import TextBlob 
```

```python
def textblob_tokenizer(input_str):
    blob = TextBlob(input_str.lower())
    tokens = blob.words
    words = [token.stem() for token in tokens]
    return words
```

```python
#Toy example: 
print(textblob_tokenizer('Q: studed studing!!! I miss uuuu! It&#039;s'))
```

```python
#countvectorizer convers each review into a vector based on the word count.
countvectorizer = CountVectorizer(analyzer= 'word', stop_words= 'english', 
                                  tokenizer=textblob_tokenizer) 
#converts text into a vector based on tf-idf weighting scheme. 
tfidfvectorizer = TfidfVectorizer(analyzer= 'word', stop_words= 'english', 
                                  tokenizer=textblob_tokenizer) 
```

```python
textFeatures
```

```python
count_matrix = countvectorizer.fit_transform(textFeatures) 
tfidf_matrix = tfidfvectorizer.fit_transform(textFeatures)
```

```python
print(tfidf_matrix) #print elements of the matrix.
```

```python
print(tfidf_matrix.shape) 
print(count_matrix.shape)
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(
    tfidf_matrix, movie['class-label'], test_size=0.3,random_state=8)
print(features_train.shape, features_test.shape, labels_train.shape, labels_test.shape)
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
from sklearn.metrics import accuracy_score
```

```python
#SVM classifier
from sklearn.svm import SVC
print("\nEvaluation for SVM \n")
svc = SVC(kernel='sigmoid', gamma=1.0)
svc.fit(features_train, labels_train)
prediction = svc.predict(features_test)
acc = accuracy_score(labels_test,prediction)
print('Accuracy:', acc)
from sklearn.metrics import precision_score
prec = precision_score(labels_test,prediction, average='weighted')
print('Precision:', prec)
from sklearn.metrics import recall_score
recall = recall_score(labels_test,prediction, average='weighted')
print('Recall:', recall)
from sklearn.metrics import f1_score
f1 = f1_score(labels_test,prediction, average='weighted')
print('F-1 measure: ', f1)
print('\nConfusion Matrix:\n')
print(confusion_matrix(labels_test, prediction))  
print(classification_report(labels_test, prediction))
#print(prediction)
```

```python
#Decision Tree
print("\nEvaluation for Decision Tree \n")
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(features_train, labels_train)
prediction = dtree.predict(features_test)
acc = accuracy_score(labels_test,prediction)
print('Accuracy: ', acc)
prec = precision_score(labels_test,prediction, average='weighted')
print('Precision: ', prec)
recall = recall_score(labels_test,prediction, average='weighted')
print('Recall: ', recall)
f1 = f1_score(labels_test,prediction, average='weighted')
print('F-1 measure: ',f1)
print('\nConfusion Matrix:\n')
print(confusion_matrix(labels_test, prediction))  
print(classification_report(labels_test, prediction))
```

```python
sid = SentimentIntensityAnalyzer()

positive_text = []
negative_text = []

for t in text:
    ss = sid.polarity_scores(t)
    if ss['compound'] >= 0.05:
        positive_text.append(t)
    elif ss['compound'] <= -0.05:
        negative_text.append(t)
```

```python
#Preparing text for Wordcloud
positive_all_text = ', '.join(t for t in positive_text)
negative_all_text = ', '.join(t for t in negative_text)

# Generate a word cloud image for positive reviews
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(positive_all_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
print("\033[1mWordCloud for Positive reviews")

# Generate a word cloud image for negative reviews
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(negative_all_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
print("\033[1mWordCloud for Negative reviews")
```

These wordclouds visually represent the most frequent words used in the positive and negative reviews, respectively. By looking at the wordclouds, we can make the following observations:

- Positive reviews: The most frequent words used in the positive reviews are 'great', 'best', 'love', 'good', 'amazing', 'excellent', 'wonderful', 'enjoyed', 'funny', and 'awesome'. This indicates that the positive reviews are generally praising the movie and its various aspects such as the acting, story, visuals, and humor.
- Negative reviews: The most frequent words used in the negative reviews are 'bad', 'worst', 'waste', 'awful', 'terrible', 'boring', 'disappointing', 'horrible', 'poor', and 'stupid'. This indicates that the negative reviews are generally criticizing the movie and its various aspects such as the acting, story, visuals, and overall quality.

Overall, the wordclouds provide a quick and easy way to visualize the sentiments expressed in the movie reviews, and can be useful for gaining insights into the strengths and weaknesses of the movie from the perspective of the audience.
